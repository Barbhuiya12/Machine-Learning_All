# **Decision Trees: An Easy Explanation**
> . What is a Decision Tree?


*   Imagine you're playing a game of "20 Questions". Each question you ask gets you closer to the answer. A Decision Tree is like that game but with data. It asks questions about the data and, based on the answers, makes decisions.

>  How Does It Work?
* Starts at the top (the root)
* Asks a question about a feature of the data
* Based on the answer, it goes down one of the branches.
* Continues asking questions and moving down branches until it reaches an answer (a leaf).

>Choosing the Right Questions (Attribute Selection Measures):

* To play "20 Questions" well, you'd want to start with broad questions that narrow down possibilities quickly. Similarly, Decision Trees use methods like Information Gain and Gini Index to decide which questions to ask and when. These methods help to pick the most informative questions.

> Advantages:

* Simple to Understand: Like a flowchart.
* Versatile: Can handle both numbers and categories.
* Visual: We can see the decision-making process

>  Disadvantages:

* Overthinking: Sometimes, the tree can make too many branches, leading to overfitting.
* Sensitive: A small change in data might change the tree a lot.
* Can be Biased: If one category appears a lot more than another.

> Pruning

* Trees can grow wild! Pruning is like giving them a trim. It helps by cutting off branches that aren't useful. This can make the tree smarter and prevent overfitting.


> Deep Dive into Terms:

* Entropy & Information Gain: Measure of randomness in data. Helps to decide which feature to split on to get a clear decision.
* Gini Index: Another way to decide which feature to split on. Aims for purity in decision making.
* Pruning: Cutting off parts of the tree that aren't very useful.

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
# Training the Decision Tree Regression model on the whole dataset
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state =0 )
regressor.fit(X, y)
```

```python
# Predicting a new result
regressor.predict([[6.5]])
```

```python
# Visualising the Decision Tree Regression results (higher resolution)
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
