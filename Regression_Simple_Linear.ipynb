**Simple Linear Regression** is a linear approach to modeling the relationship between a dependent variable and one independent variable. The goal is to find the best fit straight line that accurately predicts the output values within a range.

The Equation of the simple Linear Regression of is:



```
Y = mx + C
```
Where :

Y = Dependent Varibale

m = Slope of the line

C = Intercept

The slope and y-intercept are determined during the model training process. The slope (m) represents the change in y for a one-unit change in x, and the y-intercept (C) represents the predicted value of y when x is 0.



````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
````

>**numpy** and pandas are fundamental packages for scientific computing and data manipulation, respectively.

>**matplotlib.pyplot** is used for plotting data.

>**train_test_split** splits datasets into random train and test subsets.

>**LinearRegression** implements the linear regression algorithm.

>**mean_squared_error** calculates the mean squared error, a measure of prediction accuracy.

````python
data = pd.read_csv('Salary_Data.csv')
X = data.iloc[:, :-1].values  # Independent Variable
y = data.iloc[:, -1].values   # Dependent Variable
````

This loads the dataset and splits it into the independent and dependent variables.

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
````

This splits the dataset into training and testing subsets. 80% of the data will be used for training, and 20% will be used for testing

````python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
````

This creates a linear regression model and trains it on the training data.

````python
m = regressor.intercept_
C = regressor.coef_[0]
````

````python
x_limit = plt.xlim()
y_limit = plt.ylim()
x_text_position = x_limit[1] - (0.25 * (x_limit[1] - x_limit[0]))  # 25% from the right boundary
y_text_position = y_limit[1] - (0.1 * (y_limit[1] - y_limit[0]))   # 10% from the top boundary
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.text(x_text_position, y_text_position, f'y = {C:.2f} + {m:.2f}x', color='blue',ha='center')  # Adjust position as needed
plt.show()

````

This plots the real values from the training set in red and the predicted regression line in blue.

````python
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')  # The regression line remains the same
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
````

This plots the real values from the test set in red and the predicted regression line in blue.

````python
y_pred = regressor.predict(X_test)
````

This predicts the dependent variable values for the test set.

````python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
````

Computes and prints the mean squared error, which measures the average squared difference between the actual and predicted values.
